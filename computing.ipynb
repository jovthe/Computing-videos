# **Object Oriented Programming**

**class** --> blueprint / template for creating objects

**object** --> runtime instance of a class--> attributes + behaviour + identity

**encapsulation** --> bundling of data and methods within a class + private data are accessible via public methods

**inheritance** --> child class able to inherit all methods and properties of parent class

**polymorphism** -->occurance of something in multiple forms --> same method name, different implementations

```python
class Person:
  def __init__(self, name, age, height, weight, gender = 'Male'):
    #encapsulation
    self.__name = name
    self.__age = age
    self.__height = height
    self.__weight = weight
    self.__gender = gender

  def getName(self):
    return self.__name

  def getAge(self):
    return self.__age

  def setAge(self, newAge):
    #age can only increase
    if self.getAge() > newAge:
      print('Invalid')
    else:
      self.__age =  newAge

  def getHeight(self):
    return self.__height

  def setHeight(self, newHeight):
    self.__height = newHeight

  def getWeight(self):
    return self.__weight

  def setWeight(self, newWeight):
    self.__weight = newWeight

  def getGender(self):
    return self.__gender

  def display(self):
    print(f'Name: {self.__name}')
    print(f'Age: {self.__age}')
    print(f'Height: {self.__height}')
    print(f'Weight: {self.__weight}')
    print(f'Gender: {self.__gender}')

class Student(Person):
  def __init__(self, name, age, height, weight, classno, gender = 'Male'):
    #inheritance
    super().__init__(name, age, height, weight, gender)
    #additional attributes
    self.__class = classno
    self.__school = 'ABC school'

  def display(self):    #polymorphism
    print('Student')
    super().display() #accessing method from parent class

person1 = Person('John', 30, 170, 80)   #gender = male
student1 = Student('Alice', 20, 165, 60, 12, 'Female')

person1.display()
student1.display()

person1.setAge(31)
person1.display()

print(person1.getGender())
print(student1.getGender())
```

# **Search**

## **Linear Search**

```python
def linearsearch(data,target):    #iterative
  result=[]
  for i in range(len(data)):
    if data[i]==target:
      result.append(i)
  if result!=[]:
    return result
  else:
    return "not found"


data=[5,6,7,8,1,2,3,4,1,2,3,4,5,6,7,8]
print(linearsearch(data,4))
print(linearsearch(data,6))
print(linearsearch(data,9))
```

## **Binary Search**

**Iterative**

time complexity: O(logn)

space: O(1)

```python
def binarysearch(data,target):
  low=0
  high=len(data)-1
  while low<=high:
    mid=(low+high)//2
    if data[mid]==target:
      return mid
    elif data[mid]<target:
      low=mid+1
    else:
      high=mid-1
  return "not found"

data=[1,2,3,4,5,6,7,8]
print(binarysearch(data,3))
print(binarysearch(data,9))
```

**Recursive**

time complexity: O(logn)

space: O(logn)

```python
def binarysearch(data,target,high,low):
  if low>high:
    return "not found"
  mid=(high+low)//2
  if data[mid]==target:
    return mid
  elif data[mid]>target:
    high=mid-1
    return binarysearch(data,target,high,low)
  else:
    low=mid+1
    return binarysearch(data,target,high,low)


data=[1,2,3,4,5,6,7,8,1]
low=0
high=len(data)-1
print(binarysearch(data,5,high,low))
print(binarysearch(data,9,high,low))
```

# **Sorting**

## **Bubble sort**

Worst case: O(n^2)

Best case: O(n)

image.png

```python
def bubblesort(data):
  for i in range(len(data)):
    for j in range(len(data)-1):
      if data[j]>data[j+1]:
        data[j], data[j+1]=data[j+1], data[j]
  return data

data=[5,4,1,2,3,6,8,7]
print(bubblesort(data))
```

```python
def bubblesort(data):   #optimised
  for i in range(len(data)):
    sorted=True
    for j in range(len(data)-1):
      if data[j]>data[j+1]:
        data[j], data[j+1]=data[j+1], data[j]
      sorted=False
    if not sorted:
      break
  return data

data=[1,3,4,5,2,6,8,7]
print(bubblesort(data))
```

## **Insertion sort**

Worst case: O(n^2)

Best case: O(n)

image.png

```python
def insertionsort(data):
  for i in range(1, len(data)):
    j = i - 1
    while j >= 0 and data[j] > data[j+1]:
      data[j+1], data[j] = data[j], data[j+1]
      j -= 1
  return data

data = [4, 7, 1, 5, 2, 8, 3]
print(insertionsort(data))
```

## **Quicksort**

Worse case: O(n^2)

Best case: O(nlogn)

image.png

```python
def quicksort(data):
  if len(data)<=1:
    return data
  pivot=data[len(data)//2]
  left=[]
  middle=[]
  right=[]
  for i in data:
    if i<pivot:
      left.append(i)
    elif i==pivot:
      middle.append(i)
    else:
      right.append(i)
  return quicksort(left) + middle + quicksort(right)


data = [4, 7, 1, 5, 2, 8, 3]
print(quicksort(data))
```

## **Mergesort**

Best and Worse case: O(nlogn)

image.png


```python
def merge_sort(data):
    if len(data) <= 1:
        return data

    mid = len(data) // 2
    left_half = data[:mid]
    right_half = data[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left_half, right_half):
    i=0
    j=0
    merged_list = []
    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            merged_list.append(left_half[i])
            i += 1
        else:
            merged_list.append(right_half[j])
            j += 1

    merged_list += left_half[i:]
    merged_list += right_half[j:]

    return merged_list

# main
data = [4, 7, 1, 5, 2, 8, 3]
print(merge_sort(data))
```

```python
# merge sort optimised
def mergesort(A):
  #print("Splitting", A)
  if len(A) > 1: # split half
    mid = len(A) // 2
    left = A[:mid]
    right = A[mid:]

    mergesort(left) # recursively split left
    mergesort(right) # recursively split right

    i = 0 # left index
    j = 0 # right index
    k = 0 # result index

    while i < len(left) and j < len(right): # while left and right not empty
      if left[i] <= right[j]: # left smaller
        A[k] = left[i] # move left
        i += 1 # next left
      else: # right smaller
        A[k] = right[j] # move right
        j += 1 # next right
      k += 1 # next result

    while i < len(left): # shift remaining in left
      A[k] = left[i]
      i += 1
      k += 1

    while j < len(right): # shift remaining in right
      A[k] = right[j]
      j += 1
      k += 1

  #print("Merging ", A)

# main
data = [54, 26, 93, 17, 77, 31, 44, 55, 20]
mergesort(data)
print(data)
```
